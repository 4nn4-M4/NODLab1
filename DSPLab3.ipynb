# Digital Filter Design and Analysis: Implementing FIR and IIR filters in Python. Adaptive Filtering: Applying adaptive filtering algorithms to noise reduction.

## FIR implementation

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fs = 1000
T = 2.0
t = np.arange(0, T, 1/fs)
N = len(t)

# Clean signal
f_signal = 10  # Hz
amplitude = 1.0
clean_signal = amplitude * np.sin(2 * np.pi * f_signal * t)

# Noise
np.random.seed(42)
noise_amplitude = 0.3
noise = noise_amplitude * np.random.randn(N)
noisy_signal = clean_signal + noise

# SNR (Signal-to-Noise Ratio)
signal_power = np.mean(clean_signal**2)
noise_power = np.mean(noise**2)
SNR_input = 10 * np.log10(signal_power / noise_power)
```

```python
# Vizualization
plt.figure(figsize=(12, 6))
plt.plot(t, noisy_signal, color='orange', label=f'Noisy Signal')
plt.plot(t, clean_signal, color='blue', linewidth=2, label='Clean Signal')
plt.title("Comparison of Clean and Noisy Signals")
plt.xlabel("Time [s]")
plt.ylabel("Amplitude")
plt.xlim(0, 0.5)
plt.legend(loc='upper right')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
def fir_filter(x, b):
    """
    FIR filter implementation.
    
    Parameters:
    x : ndarray
        Input signal.
    b : ndarray
        Filter coefficients.
    
    Returns:
    y : ndarray
        Filtered output signal.
    """
    M = len(b)
    y = np.zeros(len(x))
    for n in range(M, len(x)):
        y[n] = np.dot(b, x[n-M+1:n+1][::-1])
    return y
b = [0.5, 0.5]  # FIR coefficients
x = noisy_signal
y = fir_filter(x, b)

plt.figure(figsize=(10, 6))
plt.plot(t, x, label="Noisy Signal")
plt.plot(t, y, label="Filtered Signal", linewidth=2)
plt.legend()
plt.title("FIR Filter")
plt.xlabel("Time [s]")
plt.ylabel("Amplitude")
plt.grid()
plt.show()

error_fir = clean_signal - y
noise_power_fir = np.mean(error_fir**2)
SNR_fir = 10 * np.log10(signal_power / noise_power_fir)

print(f"\nFIR results:")
print(f"SNR in: {SNR_input:.2f} dB")
print(f"SNR out: {SNR_fir:.2f} dB")
print(f"SNR improvement: {SNR_fir - SNR_input:.2f} dB")
print(f"MSE: {np.mean(error_fir**2):.6f}")
```

## IIR Filter implementation

```python
import numpy as np
import matplotlib.pyplot as plt

def iir_filter(x, b, a):
    """
    IIR filter implementation.

    Parameters:
    x : ndarray
        Input signal.
    b : ndarray
        Numerator coefficients.
    a : ndarray
        Denominator coefficients.

    Returns:
    y : ndarray
        Filtered output signal.
    """
    M = len(b)  # Length of numerator coefficients (b)
    N = len(a)  # Length of denominator coefficients (a)
    y = np.zeros(len(x))  # Initialize output signal array

    # Apply filter to each sample in the input signal
    for n in range(len(x)):
        # Numerator part (feedforward)
        # Ensure we use the correct slice length for the convolution
        x_slice = x[max(0, n-M+1):n+1]  # Input signal slice
        y[n] = np.dot(b[:len(x_slice)], x_slice[::-1])  # Apply reverse convolution for numerator

        # Denominator part (feedback), skip the first sample
        if n >= 1:
            # Ensure we use the correct slice length for the feedback part
            y_slice = y[max(0, n-N+1):n]  # Output signal slice
            y[n] -= np.dot(a[1:min(N, len(y_slice)+1)], y_slice[::-1])  # Apply reverse convolution for feedback

    return y

# Example usage and plotting
# Create a noisy signal (for example, a sine wave with noise)
# fs = 1000  # Sampling frequency
# t = np.linspace(0, 1, fs)  # Time vector
# x = np.sin(2 * np.pi * 50 * t) + 0.5 * np.random.randn(len(t))  # Noisy signal

# IIR filter coefficients
a = [1, -0.4]  # Denominator coefficients (a_0 = 1 by convention)
b = [0.3, 0.4]  # Numerator coefficients

# Apply the filter to the noisy signal
y = iir_filter(x, b, a)

# Plot the results
plt.figure(figsize=(10, 6))
plt.plot(t, x, label="Noisy Signal")
plt.plot(t, y, label="Filtered Signal", linewidth=2)
plt.legend()
plt.title("IIR Filter Response")
plt.xlabel("Time [s]")
plt.ylabel("Amplitude")
plt.grid(True)
plt.show()

error_iir = clean_signal - y
noise_power_iir = np.mean(error_iir**2)
SNR_iir = 10 * np.log10(signal_power / noise_power_iir)

print(f"\nIIR results:")
print(f"SNR in: {SNR_input:.2f} dB")
print(f"SNR out: {SNR_iir:.2f} dB")
print(f"SNR improvement: {SNR_iir - SNR_input:.2f} dB")
print(f"MSE: {np.mean(error_iir**2):.6f}")
```

## Adaptive LMS Filter Implementation

```python
def lms_filter(x, d, mu, num_taps):
    """
    LMS adaptive filter implementation.
    
    Parameters:
    x : ndarray
        Input signal (noisy).
    d : ndarray
        Desired signal.
    mu : float
        Step size.
    num_taps : int
        Number of filter taps.
    
    Returns:
    y : ndarray
        Filtered output signal.
    e : ndarray
        Error signal.
    w : ndarray
        Final filter weights.
    """
    n = len(x)
    w = np.zeros(num_taps)
    y = np.zeros(n)
    e = np.zeros(n)

    for i in range(num_taps, n):
        x_segment = x[i-num_taps:i][::-1]
        y[i] = np.dot(w, x_segment)
        e[i] = d[i] - y[i]
        w += mu * e[i] * x_segment

    return y, e, w

# Example usage and plotting
# d = np.sin(2 * np.pi * 5 * t)  # Desired signal
d = clean_signal
mu = 0.1  # Step size
num_taps = 3

y, e, w = lms_filter(x, d, mu, num_taps)

plt.figure(figsize=(10, 6))
plt.plot(t, d, label="Desired Signal")
plt.plot(t, y, label="Filtered Signal", linewidth=2)
plt.legend()
plt.title("LMS Adaptive Filter")
plt.xlabel("Time [s]")
plt.ylabel("Amplitude")
plt.grid()
plt.show()

noise_power_lms = np.mean(e**2)
SNR_lms = 10 * np.log10(signal_power / noise_power_lms)

print(f"\nLMS results:")
print(f"SNR in: {SNR_input:.2f} dB")
print(f"SNR out: {SNR_lms:.2f} dB")
print(f"SNR improvement: {SNR_lms - SNR_input:.2f} dB")
print(f"MSE: {np.mean(e**2):.6f}")
```
