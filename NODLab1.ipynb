Wyświetlić wykres na podstawie tabeli przystawnej

```python
df[(df['risk_name'] == 'High body-mass index') &
  (df['sex'] == 'Female')].pivot_table(values='mean', index='acause', columns='risk_name', aggfunc='mean',
                                          fill_value=None, margins=False, dropna=True).plot(kind = 'line')

plt.ylabel('Mean estimate') 
plt.title('Expenses for the High body-mass index with cause problems considered')
```

Narysować histogram na podstawie wartości kolumny

```python
df_bar = df[df['sex'].isin(['Male','Female'])].pivot_table(values='mean', 
                index='risk_name', columns='sex', aggfunc='mean',
                      fill_value=None, margins=False, dropna=True)
df_bar.plot(kind = 'bar')
plt.ylabel('Mean estimate')
plt.title('Expenses for health by the risk factors with consideration of gender ')
```

Przedstawić sposoby  łączenia ramek danych za pomocą metod merge i
concat

```python
df1=df[df["cause_name"] == "Cardiovascular diseases"] 
df1.reset_index(drop=True, inplace=True)
df2=df[df["cause_name"] == "Communicable diseases"] 
df2.reset_index(drop=True, inplace=True)

df_all = pd.merge(df1, df2, on = ['location_id','location_name', 'year', 'sex_id', 'sex', 'age_group_id', 'age_group_name', 'risk_id','risk_name', 'metric'], how = 'inner')
df_all
```

```python
df_all_new = pd.concat([df1, df2], axis = 0)
df_all_new
```

Pokazać dodawanie nowych kolumn za pomocą operacji matematycznych

```python
df["mean_round"] = df["mean"].round(decimals = 2)
df["mean_total"] = df["mean"]+df["lower"]+df["upper"]
df["tmean_med"] = df.loc[:, "mean":"upper"].median(axis = 1)
df 
```

Przedstawić na przykładzie dodawanie nowych kolumn z pomocą funkcji

```python
youngest = ['20 to 44', '<20 years', '20 to 32', '32 to 44']
df['youngest'] = df['age_group_name'].apply(lambda x: True if x in youngest else False )
df[df['youngest'] == True]
```
